```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from MDParser import AnalysisParser

sns.set_style('darkgrid')
```

```python
def plotForDenoiser(analysisData, variable, denoiser):

    sigmas = list(sorted(analysisData[0].keys()))
    values = []
    for sigma in sigmas:
        if denoiser == 'KF' and sigma==3:
            values.append(analysisData[0][sigma-0.1][denoiser][variable])
        else:
            values.append(analysisData[0][sigma][denoiser][variable])

    plt.plot(sigmas, values)
    plt.xlabel("Ïƒ")
    plt.xticks(sigmas, rotation=90)
    plt.ylabel(variable)
```

```python
parser = AnalysisParser()
```

### Model 4

```python
analysisDataModel1 = parser.parseAnalysisData('../SBAgent/models/base/model4/best_model.zip_eval.md_unbiased.md')

plotForDenoiser(analysisDataModel1, 'Success Rate', 'None')
plotForDenoiser(analysisDataModel1, 'Success Rate', 'LPF')
plotForDenoiser(analysisDataModel1, 'Success Rate', 'KF')
plt.legend(["No Denoiser", "LPF", "KF"], frameon=False)
plt.ylim([0,1])
plt.xticks(np.arange(0,3.1,0.2))
plt.show()

plotForDenoiser(analysisDataModel1, 'Mean Reward', 'None')
plotForDenoiser(analysisDataModel1, 'Mean Reward', 'LPF')
plotForDenoiser(analysisDataModel1, 'Mean Reward', 'KF')
plt.legend(["No Denoiser", "LPF", "KF"], frameon=False)
plt.ylim([-6000,-500])
plt.show()
```
