```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from MDParser import AnalysisParser

sns.set_style('darkgrid')
```

```python
def plotForDenoiser(analysisData, variable, denoiser):

    mus = list(sorted(analysisData.keys()))
    values = []
    for mu in mus:
        values.append(analysisData[mu][0][denoiser][variable])

    plt.plot(mus, values)
    plt.xlabel("Î¼")
    plt.xticks(mus, rotation=90)
    plt.ylabel(variable)
```

```python
parser = AnalysisParser()
```

### Model 1

```python
analysisDataModel1 = parser.parseAnalysisData('../SBAgent/evalResults/model1/biased.md')

plotForDenoiser(analysisDataModel1, 'Success Rate', 'None')
plotForDenoiser(analysisDataModel1, 'Success Rate', 'LPF')
plotForDenoiser(analysisDataModel1, 'Success Rate', 'KF')
plt.legend(["No Denoiser", "LPF", "KF"], frameon=False)
plt.ylim(0,1)
plt.xticks(np.arange(0,0.31,0.02))
plt.show()

plt.title("Mean Reward for Model1 (No Noise During Training)")
plotForDenoiser(analysisDataModel1, 'Mean Reward', 'None')
plotForDenoiser(analysisDataModel1, 'Mean Reward', 'LPF')
plotForDenoiser(analysisDataModel1, 'Mean Reward', 'KF')
plt.legend(["No Denoiser", "LPF", "KF"], frameon=False)
plt.show()
```
