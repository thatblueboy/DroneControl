```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Reward**

```python
model1Reward = pd.read_csv('trainData/model4-rew.csv')
model2Reward = pd.read_csv('trainData/model5-rew.csv')
model3Reward = pd.read_csv('trainData/model6-rew.csv')
model4Reward = pd.read_csv('trainData/model10_reward.csv')
model5Reward = pd.read_csv('trainData/model11_reward.csv')
model6Reward = pd.read_csv('trainData/model12_reward.csv')
```

```python
rewardData1 = []
rewardData2 = []

for r1, r2, r3 in zip(model1Reward.iterrows(), model2Reward.iterrows(), model3Reward.iterrows()):
    rewardData1.append([r1[1]['Step'], r1[1]['Value'], r2[1]['Value'], r3[1]['Value']])
    
for r4, r5, r6 in zip(model4Reward.iterrows(), model5Reward.iterrows(), model6Reward.iterrows()):
    rewardData2.append([r4[1]['Step'], r4[1]['Value'], r5[1]['Value'], r6[1]['Value']])
# rewardData2.append([r4[1]['Step'], r4[1]['Value'], r5[1]['Value'], r6[1]['Value']])
rewardData1= np.array(rewardData1)
rewardData2= np.array(rewardData2)
print(rewardData1)
print(rewardData2)
```

```python
sns.set_style('darkgrid')
sns.despine()

maxStep = 1000

# plt.plot(rewardData1[:,0][:maxStep], rewardData1[:,1][:maxStep])
# plt.plot(rewardData1[:,0][:maxStep], rewardData1[:,2][:maxStep])
# plt.plot(rewardData1[:,0][:maxStep], rewardData1[:,3][:maxStep])
plt.plot(rewardData2[:,0][:maxStep], rewardData2[:,1][:maxStep])
plt.plot(rewardData2[:,0][:maxStep], rewardData2[:,2][:maxStep])
plt.plot(rewardData2[:,0][:maxStep], rewardData2[:,3][:maxStep])
plt.legend(['Policy 1, Environment 1', 'Policy 2, Environment 1', 'Policy 3, Environment 1'], frameon=False)
plt.xlabel("Timestep")
plt.ylabel("Mean Reward")
plt.savefig("TrainRewardVaryingBias.png")
```
